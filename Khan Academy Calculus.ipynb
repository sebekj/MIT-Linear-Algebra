## Idea of a Limit
Examples of functions with limits:

$$
\begin{align}
f(x) &= \frac{x-1}{x-1} \\
\lim_{x \to 1} f(x) &= 1 \\ \\
g(x) &= 
\begin{cases}
x^2, & \text{if } x \neq 2 \\
1,   & \text{if } x = 2
\end{cases} \\
\lim_{x \to 2} g(x) &= 4
\end{align}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Create x values, avoid x=2 for the curve
x = np.linspace(0, 4, 400)
x_curve = x[x != 2]  # Remove x=2 to show discontinuity

# Define g(x)
g = lambda x: x**2

# Plot x^2 except at x=2
plt.plot(x_curve, g(x_curve), label=r'$g(x) = x^2,\, x\neq 2$')

# Hollow circle at (2, 4) to show the 'missing' value
plt.plot(2, 4, 'ro', mfc='none', markersize=10, label=r"open circle at $(2, 4)$")

# Solid circle at (2, 1)
plt.plot(2, 1, 'ro', label=r"$g(2) = 1$")

# Add details
plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('Piecewise Function $g(x)$')
plt.legend()
plt.grid(True)
plt.ylim(0, 6)

plt.show()
```

Note:

* Limits can be defined by the function, or not.
* The limit does not exist when a limit does not approach the same value from both sides.
* **Unbounded limits** do not exist, such as

$$
\begin{align}
y &= \frac{1}{x^2} \\
\lim_{x \to 0} \frac{1}{x^2} &= \text{ Does not exist.}
\end{align}
$$

* It is possible for a function value to be different from the limit value.
* One-sided limites from graphs can be noted as,

$$
\begin{align}
\lim_{x \to 2^{-}} f(x) &= 5 \\
\lim_{x \to 2^{+}} f(x) &= 1 \\
\lim_{x \to 2} f(x) & \text{ does not exist}
\end{align}
$$

## Standard Definition of Limits $\lim_{x \to c} f(x) = L$.

Given $\epsilon > 0$ we can find $\delta > 0$ such that if $|x - c| < \delta$ then $|f(x) - L| < \epsilon$.



## Properties of Limits

$
\begin{align}
\text{When } 
\lim_{x \to c} f(x) = L &\text{ and } \lim_{x \to c} g(x) = M \text{ then } \\ \\
\lim_{x \to c} (f(x) + g(x)) &= L + M \\ \\
\text{Similarly:} \\
\lim_{x \to c} (f(x) - g(x)) &= L - M \\
\lim_{x \to c} (f(x) \times g(x)) &= L \times M \\
\lim_{x \to c} k(f(x)) &= kL
\end{align}
$

## Limits of Composite Functions

$$
\lim_{x \to a} f(g(x)) = f\left(\lim_{x \to a} g(x)\right) \iff 
\begin{cases}
\lim_{x \to a} g(x) = L \quad \text{(i.e., the limit exists)} \\
f \text{ is continuous at } L
\end{cases}
$$

Still, in special cases when this theorem cannot be applied, the limit of $f(x)$ can be determined.

$$

\end{align}

So that function $g(x)$ can be reduced:

This can be evaluated at $x = -1$

$$

## Cosine and Sine identities etc.

The general sum formula for sine and for cosine:

$$
\begin{align} \sin(a + b) &= \sin(a)\cos(b) + \cos(a)\sin(b) \\ \cos(a + b) &= \cos(a)\cos(b) - \sin(a)\sin(b) \end{align}
$$

Note that for further substitutions you can take â€Š$a = b = x$â€Š to achieve $\sin(2x)$â€Š and $\cos(2x)$â€Š.

```python
import numpy as np
import matplotlib.pyplot as plt

# Example angle (change to experiment)
x = np.pi / 6  # 30 degrees

# Unit circle
theta = np.linspace(0, 2 * np.pi, 400)
circle_x = np.cos(theta)
circle_y = np.sin(theta)

plt.figure(figsize=(7,7))
plt.plot(circle_x, circle_y, label='Unit Circle')
plt.axhline(0, color='gray', lw=0.8)
plt.axvline(0, color='gray', lw=0.8)

# Point at angle x
Px, Py = np.cos(x), np.sin(x)
plt.plot(Px, Py, 'bo', label=r'$(\cos(x), \sin(x))$')

# Line from origin to (cos(x), sin(x))
plt.plot([0, Px], [0, Py], 'b--')

# Show cos(x)
plt.plot([0, Px], [0, 0], 'g-', lw=2, label=r'$\cos(x)$')
plt.text(Px/2, -0.08, r'$\cos(x)$', color='g', ha='center')

# Show sin(x)
plt.plot([Px, Px], [0, Py], 'm-', lw=2, label=r'$\sin(x)$')
plt.text(Px+0.07, Py/2, r'$\sin(x)$', color='m')

# Area rectangle for sin(x)*cos(x) (for positive x and y only)
rect_x = [0, Px, Px, 0, 0]
rect_y = [0, 0, Py, Py, 0]
plt.fill(rect_x, rect_y, color='orange', alpha=0.2, label=r'$\sin(x)\cos(x)$')

# Show point at angle 2x
Qx, Qy = np.cos(2*x), np.sin(2*x)
plt.plot(Qx, Qy, 'ro', label=r'$(\cos(2x), \sin(2x))$')
plt.plot([0, Qx], [0, Qy], 'r--')

# Lines to y-axis showing sin(2x) and 2*sin(x)*cos(x)
plt.plot([Qx, Qx], [0, Qy], 'r:', lw=2)
plt.text(Qx+0.07, Qy/2, r'$\sin(2x)$', color='r')

# Numerically show both values
sin2x = np.sin(2*x)
double_sin_x_cos_x = 2 * np.sin(x) * np.cos(x)

plt.title(r'$\sin(2x)$ vs $2\sin(x)\cos(x)$ (unit circle, $x = 30^\circ$)')
plt.axis('equal')
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend(loc='lower left')
plt.grid(True)

# Text summary
plt.text(-1.1, 1.05, rf'$\sin(2x) = {sin2x:.3f}$', fontsize=12, color='r')
plt.text(-1.1, 0.9, rf'$2\sin(x)\cos(x) = {double_sin_x_cos_x:.3f}$', fontsize=12, color='orange')

plt.show()
```

### Definition of Continuity

The $f$ is continuous at $x = c$ when $\lim_{x \to c} f(x) = f(c)$.

### Continuity over an Internal

The $f$ is continuous over an open interval $(a, b)$ if and only if $f$ is continuous over every point in the interval.

The $f$ is continuous over a closed interval $[a, b]$ if and only if $f$ is continuous over $(a, b)$  and $\lim_{x \to a^+} f(x) = f(a)$ and $\lim_{x \to b^-} f(x) = f(b)$.

### Infinite limits

$y = \frac{1}{x^2}, \lim_{x \to 0} \frac{1}{x^2} = \infty$

### Intermediate Value Theorem

Suppose $f$ is a function continuous at every poinit at the closed interval $[a, b]$:

* $f$ will take on every value between $f(a)$ and $f(b)$ over the interval.
* For any $L$ between the vallues $f(a)$ and $f(b)$, there exists at least one number $c$ in $[a, b]$ for which $f(c) = L$.

## Differential Calculus

**Slope: rise over run**

$$\text{slope } = \frac{\Delta y}{\Delta x}$$

Differential Calculus determines the instantaneous rate of change which is the tangent line. The slope of the tangent line at a point, $\frac{\delta y}{\delta x}$ or $f'$ (*f prime*) or $\dot{y}$, is the instantaneous rate of change.

Example: $y = x^2$. What is the average rate of change of $y$ with respect of $x$, over $[1, 3]$

1. determine the values

| $x$ | $y = x^2$|
|-----|----------|
|    1|         1|
|    3|         9|

Here, $\Delta x = 2$ and $\Delta y = 8$. The average rate of change is $\frac{\Delta y}{\Delta x} = \frac{8}{2} = 4$.

This is the slope of the *secant* line.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Values
x_vals = np.array([1, 3])
y_vals = x_vals**2

# Secant line slope (average rate of change)
delta_x = x_vals[1] - x_vals[0]
delta_y = y_vals[1] - y_vals[0]
slope = delta_y / delta_x

# Equation of secant line: y = m(x - x0) + y0
def secant_line(x):
    return slope * (x - x_vals[0]) + y_vals[0]

# x for plot
x = np.linspace(0, 4, 200)
y = x**2

plt.figure(figsize=(7,5))
plt.plot(x, y, label=r'$y = x^2$', lw=2)
plt.scatter(x_vals, y_vals, color='red', zorder=5)
plt.plot(x, secant_line(x), 'g--', label='Secant line', lw=2)

plt.text(1, 1, r'$(1, 1)$', ha='right', va='top', color='red')
plt.text(3, 9, r'$(3, 9)$', ha='left', va='bottom', color='red')
plt.title('Secant Line for $y = x^2$ from $x=1$ to $x=3$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Show the table
import pandas as pd
df = pd.DataFrame({'$x$': x_vals, '$y = x^2$': y_vals})
display(df)

# Print the average rate of change
print(f"Î”x = {delta_x}")
print(f"Î”y = {delta_y}")
print(f"Average rate of change: Î”y/Î”x = {slope}")
```

### Derivative Notations

1. Lagrange's

The derivative of $f$ is expressed as $f'$, pronounced 'f prime'.

2. Leibnitz's

The derivative of $f$ is expressed as $\frac{d}{dx}f(x)$. In case of an equation $y = f(x)$ we can express the derivate as $\frac{dy}{dx}$.

3. Newton's

The derivative of $f$ is epxressed as $\dot{f}$, and the derivative of $y = f(x)$ as $\dot{y}$.

## Definition of a Derivative

1. Generalise the slope of a secant between two points with a distance $h$ between them.

Known points are, $(x_0, f(x_0))$ and $(x_0+h, f(x_0+h))$.

The slope is,

$$
\frac{\Delta y}{\Delta x} = \frac{f(x_0+h) - f(x_0)}{x_0+1-x_0} = \frac{f(x_0+h) - f(x_0)}{h}
$$

2. Then, the slope of the tangent is at $h = 0$, it is this $f$ prime:

$$
f'(x) = \lim_{h \to 0} \frac{f(x_0+h) - f(x_0)}{h}
$$

3. The alternate form of the derivative,

Take two points, $(x, f(x))$ and $(a, f(a)$ and find the limit to,

$$
\lim_{x \to a} \frac{f(x) - f(a)}{x-a}
$$

4. A worked example

For function $y = x^2$ determine the slope of the secant between point $3$ and a point $\Delta x$ away from it. Using a limit, determine $f'(3)$.

| $x$ | $y = x^2$|
|-----|----------|
|  $3$|     $9$|
|$3+\Delta x$|$(3+\Delta x)^2$|

Then, the slope of the secant between these two points is,

$$
\frac{\Delta y}{\Delta x} =
\frac{(3+\Delta x)^2 - 9}
{3+\Delta x - 3} =
\frac{9 + 6\Delta x + (\Delta x)^2 - 9}
{\Delta x} = 
6 + \Delta x
$$

Then, determine the limit at point $3$ when the distance $\Delta x$ is $0$,

$$
\lim_{\Delta x \to 0} 6 + \Delta x = 6
$$

So, for function $y = x^2$, at point $3$, the slope of the tangent is $f'(3) = 6$.

5. A general formula,

For function $y = x^2$ determine the slope of the secant between point $x$ and a point $+ h$ away from it. Using a limit, determine $f'(x)$ for any point.

| $x$ | $y = x^2$|
|-----|----------|
|  $x$|   $x^2$|
|$x+ h$|$(x+ h)^2$|

Then, the slope of the secant between these two points is,

$$
\frac{\Delta y}{\Delta x} =
\frac{(x + h)^2 - x^2}
{h} =
\frac{x^2 + 2xh + h^2 - x^2}
{h} = 2x + h
$$

Then, determine the limit as the distance $h$ narrows to $0$:

$$
\lim_{h \to 0} = 2x + h = 2x
$$

Then, at any $x$ for function $f(x) = x^2$, the slope of the tangent is $f'(x) = 2x$.

## Differentiability and continuity

For a general differential,

$$
f'(c) = \lim_{x \to c} \frac{f(x) - f(c)}{x - c}
$$

* If $f$ is differentiable at $x = c$ then $f$ is continuous at $x = c$.
* If $f$ is not continuous at $x = c$ then $f$ is not differentiable at $x = c$.

Note that e.g. continous function $y = |x - c|$ is not differentiable at $c$ even though it is continuous. See here:

```python
import numpy as np
import matplotlib.pyplot as plt

c = 2  # Change this value for a different "corner" point
x = np.linspace(c - 4, c + 4, 400)
y = np.abs(x - c)

plt.figure(figsize=(7,5))
plt.plot(x, y, label=fr'$y = |x - {c}|$', color='blue', lw=2)
plt.scatter([c], [0], color='red', zorder=5)
plt.text(c, 0.3, fr'Not differentiable at $x = {c}$', color='red', ha='center')
plt.axvline(c, color='gray', linestyle='--', lw=1)
plt.title(fr'Graph of $y = |x - {c}|$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

Note that differentiability implies continuity:

$$
\lim_{x \to c} f(x) = f(c)
$$

## The power rule 

The definition of a derivative:

$$
\lim_{\Delta x \to 0} \frac{f(x+\Delta x) - f(x)}{\Delta x}
$$

This rule proves that,

$$
\begin{align}
f(x) & = x^n, n \neq 0 \\ 
f'(x) & = nx^{n-1}
\end{align}
$$

## Derivative of $a^x$ for any positive base $a$

When exploring the property of $\frac{d}{dx}e^x = e^x$ then we can evaluate,

$$
\frac{d}{dx}a^x
$$

By using the logarithmic property which is true for **any positive real number**,

$$
a = e^{\ln(a)}
$$

Which answers $ln(a)$, that is the question '**to what power must I raise $e$ so that I get $a$**?

Then, by using exponent properties, from this it follows,

$$
\frac{d}{dx}[a^x] = \frac{d}{dx}[(e^{\ln(a)})^x] = \frac{d}{dx}[e^{(\ln(a))x}]
$$

Now, we can use the chain rule to evaluate this,

$$
\begin{align}
\frac{d}{dx}[\underbrace{e^{\overbrace{(\ln(a))x}^{c_1(x)}}}_{c_2(x)}] & =
\frac{d}{dx}c_2(c_1(x)) = \underbrace{e^{\ln(a)x}}_{c'_2}
\cdot \underbrace{\ln(a)}_{c'_1} =
\\
& = (\ln(a))(\underbrace{e^{\ln(a)}}_{a})^x = \ln(a)a^x
\end{align}
$$

In conclusion,

$$
\frac{d}{dx}a^x = \ln(a)a^x
$$

## Derivative of $\log_a x$ for any positive base $a \neq 1$

Let us evaluate,

$$
\frac{d}{dx} [\log_a x]
$$

This is the change of base,

$$
\log_a b = \frac{\log_c b}{\log_c a}
$$

Let us leverage the natural log,

$$
\frac{d}{dx} [\log_a x] = \frac{d}{dx} [\frac{\ln x}{\ln a}]
$$

Since, $\ln a$ is just a number, let us rewrite to,

$$
\frac{d}{dx} [\log_a x] = \frac{d}{dx} [\underbrace{\frac{1}{\ln a}}_{\text{a number}} \cdot \ln x] = 0 \cdot \ln x + \frac{1}{\ln a} \cdot \underbrace{\frac{1}{x}}_{(\ln x)'}
$$

In conclusion,

$$
\frac{d}{dx} [\log_a x] = \frac{1}{(\ln a) x}
$$

## Implicit Differentiation

Let us determine the slope of the unit circle, without explicitly splitting it into two functions $y = \sqrt(1 - x^2)$ and $y = -\sqrt(1-x^2)$, and differentiating these two functions separately.

We start off with the unit circle equation:

$$
x^2 + y^2 = 1
$$

We differentiate both sides with respect to $x$:

$$
\begin{align}
\frac{d}{dx}\left[x^2 + y^2\right] &= \frac{d}{dx}\left[ 1 \right] \\
\underbrace{\frac{d}{dx} \left[ x^2 \right]}_{2x} +
\underbrace{\frac{d}{dx} \left[ y^2 \right]}_{\text{apply chain rule}} &= 0 \\
\underbrace{\frac{d}{dx} \left[ x^2 \right]}_{2x} +
\underbrace{\frac{d}{dx} \left[ y^2 \right]}_{\frac{d(y^2)}{dy} \cdot \frac{dy}{dx}} &= 0 \\
2x + 2y \cdot
\underbrace{\frac{dy}{dx}}_{\text{slope of tangent line}} &= 0 \\
\frac{dy}{dx} = -\frac{2x}{2y} = -\frac{x}{y}
\end{align}
$$

Say, we want to find the slope of the tange line of a point at $45^{\circ}$ angle, $(\frac{\sqrt{2}}{2}, \frac{\sqrt{2}}{2})$:

$$
\begin{align}
\frac{dy}{dx} &= -\frac{x}{y} \\
&= -\frac{\frac{\sqrt{2}}{2}}{\frac{\sqrt{2}}{2}} \\
\frac{dy}{dx} & = -1
\end{align}
$$

### Differentiation with Respect to $x$: An advanced example

Letâ€™s implicitly differentiate $(x^2 + y^2)^3 = 5x^2 y^2$ with respect to $x$:

**Left Side (Chain Rule)**

Let $u = x^2 + y^2$:

$
\frac{d}{dx}(u^3) = 3u^2 \cdot \frac{du}{dx}
$

But $\frac{du}{dx} = 2x + 2y \frac{dy}{dx}$:

So,

$
\frac{d}{dx}(x^2 + y^2)^3 = 3(x^2 + y^2)^2 \cdot (2x + 2y \frac{dy}{dx})
$

**Right Side (Product Rule)**

$
\frac{d}{dx}(5x^2 y^2) = 5\left[2x y^2 + x^2 \cdot 2y \frac{dy}{dx}\right]
= 10x y^2 + 10x^2 y \frac{dy}{dx}
$

Put Together and Solve for $\frac{dy}{dx}$:

$
3(x^2 + y^2)^2 (2x + 2y \frac{dy}{dx}) = 10x y^2 + 10x^2 y \frac{dy}{dx}
$

Expand left:

$
6x (x^2 + y^2)^2 + 6y (x^2 + y^2)^2 \frac{dy}{dx} = 10x y^2 + 10x^2 y \frac{dy}{dx}
$

Gather all $\frac{dy}{dx}$ terms on one side:

$
6y (x^2 + y^2)^2 \frac{dy}{dx} - 10x^2 y \frac{dy}{dx} = 10x y^2 - 6x (x^2 + y^2)^2
\left[6y (x^2 + y^2)^2 - 10x^2 y\right] \frac{dy}{dx} = 10x y^2 - 6x (x^2 + y^2)^2
$

$
\frac{dy}{dx} = \frac{10x y^2 - 6x (x^2 + y^2)^2}{6y (x^2 + y^2)^2 - 10x^2 y}
$

Final Answer:
$
\boxed{
\frac{dy}{dx} = \frac{10x y^2 - 6x(x^2 + y^2)^2}{6y(x^2 + y^2)^2 - 10x^2 y}
}

## Chain Rule and Inverse Functions

For a function $f(x)$, the inverse is

$
g(x) = f^{-1}(x)
$

Then,

$
g(f(x)) = x
$

Taking the derivative with respect to $x$:

$
\frac{d}{dx} g(f(x)) = \frac{d}{dx} x
$

On right handside we apply the *chain rule*:

$
g'(f(x)) \cdot f'(x) = 1
$

Then, we get a relationship of derivatives of inverse functions:

$
\boxed{
f'(x) = \frac{1}{g'(f(x))}
}
$

## Derivative of inverse $\sin$ and $\cos$ functions

When,

$
y = sin^{-1}(x)
$

What is the derivative with respect to $x$, $\frac{dy}{dx}$ while the inverse of $\sin$ is not known?

We start taking the derivative with respect to $x$ from both sides of the equation for the known inverse of function $y = \sin(x)$ which is $x = \sin(y)$

$
\frac{d}{dx}\left[ \sin(y) \right] = \frac{d}{dx}\left[ x \right]
$

Which is, after applying the *chain rule*,

$
\left( \cos(y) \right) \cdot \frac{dy}{dx} = 1
$

$
\frac{dy}{dx} = \frac{1}{\cos(y)}
$

With respect to the entity $\sin^2(x) + \cos^2(x) = 1$ resulting in $\cos(y) = \sqrt{1 - \sin^2(y)}$,

$
\frac{dy}{dx} = \frac{1}{\sqrt{1 - \sin^2(y)}}
$

Now, for function $y = \sin(x)$, the inverse is $x = \sin(y)$,

$
\frac{dy}{dx} = \frac{1}{\sqrt{1 - x^2}}
$

So, the derivative of the inverse is,

$
\boxed{
\frac{d}{dx} \left[ \sin^{-1}(x) \right] =  \frac{1}{\sqrt{1 - x^2}}
}
$

Similarly,

The derivative with respect to $x$ of this inverse function,

$
y = \cos^{-1}(x)
$

is,

$
\boxed{
\frac{d}{dx} \left[ \cos^{-1}(x) \right] =  - \frac{1}{\sqrt{1 - x^2}}
}
$

## Composite exponential function differentiation

Find the derivative of $y = x^x$.

Take $\ln$ of both sides of the equation,

$

Find the derivative,

## L'HÃ´pital's Rule

Evaluating limits which, when evaluated, end up in indeterminate, undefined forms.

**Case 1 for indeterminate form $\frac{0}{0}$:**

For $\lim_{x\to c} f(x) = 0 \quad \text{and} \quad
\lim_{x\to c} g(x) = 0 \quad \text{and} \quad
\lim_{x\to c} \frac{f'(x)}{g'(x)} = L \quad \text{then} \quad
\boxed{
\lim_{x\to c} \frac{f(x)}{g(x)} = L
}$

**Case 2 for indeterminate form $\frac{\pm \infty}{\pm \infty}$:**

For $\lim_{x\to c} f(x) = \pm \infty \quad \text{and} \quad
\lim_{x\to c} g(x) = \pm \infty \quad \text{and} \quad
\lim_{x\to c} \frac{f'(x)}{g'(x)} = L \quad \text{then} \quad
\boxed{
\lim_{x\to c} \frac{f(x)}{g(x)} = L
}$

Note: L'HÃ´pital's Rule can be applied again if the result is an interdeterminate form.

## Example of 10 derivations required before the limit is determinate

Letâ€™s consider:

$
L = \lim_{x \to 0} \frac{x^{10}}{e^x - 1 - x - \frac{x^2}{2!} - \cdots - \frac{x^9}{9!}}
$

Step 1: Understand the Problem

* The denominator is the Taylor expansion of $e^x$ at $0$, with all terms up to $x^9$ removed:

$
e^x = 1 + x + \frac{x^2}{2!} + \cdots + \frac{x^{10}}{10!} + \cdots
$

* Thus, the denominator behaves like $\frac{x^{10}}{10!} + \cdots$ as $x \to 0$.

Step 2: First few derivatives

Numerator: $f(x) = x^{10}$
Denominator: $g(x) = e^x - 1 - x - \frac{x^2}{2!} - \cdots - \frac{x^9}{9!}$

For all $k < 10$,
* $f^{(k)}(0) = 0$
* $g^{(k)}(0) = 0$

So, after each of the first 9 derivatives, you still get $\frac{0}{0}$.

Step 3: The 10th Derivative

Numerator:

$
\frac{d^{10}}{dx^{10}} x^{10} = 10!
$

Denominator:

$
\frac{d^{10}}{dx^{10}} \left[ e^x - 1 - x - \frac{x^2}{2!} - \cdots - \frac{x^9}{9!} \right]
$

Letâ€™s look at each term:
* $e^x$: 10th derivative is $e^x$, which is $1$ at $x=0$
* All other terms ($1, x, x^2, \ldots, x^9$): their 10th derivatives are $0$.

So, at $x = 0$:

$
\frac{d^{10}}{dx^{10}} g(x)\Big|_{x=0} = 1
$

Step 4: Apply Lâ€™HÃ´pitalâ€™s Rule 10 times

Each time you apply Lâ€™HÃ´pitalâ€™s Rule, you differentiate numerator and denominator:

$
L = \lim_{x \to 0} \frac{x^{10}}{g(x)} =
\lim_{x \to 0} \frac{f^{(10)}(x)}{g^{(10)}(x)} =
\frac{10!}{1} = 3628800
$

Step 5: Final Answer

$
\boxed{
\lim_{x \to 0} \frac{x^{10}}{e^x - 1 - x - \frac{x^2}{2!} - \cdots - \frac{x^9}{9!}} = 10!
}
$
Or numerically:â€ƒ$\boxed{3628800}$

Summary
* This example requires 10 applications of Lâ€™HÃ´pitalâ€™s Rule, since both numerator and denominator and their first 9 derivatives at 0 vanish.
* Only the 10th derivative yields a nonzero numerator and denominator.

## Mean Value Theorem

Let $f$ be a function that is continuous on the closed interval $[a, b]$ and differentiable on the open interval $(a, b)$.

The **average rate of change** of $f$ between $a$ and $b$ is the slope of the *secant* line:

$
\frac{\Delta y}{\Delta x} = \frac{f(b)-f(a)}{b-a}.
$

The **Mean Value Theorem** states that there exists a point $c \in (a, b)$ such that the **instantaneous rate of change** (the derivative) at $c$ equals this average rate of change:

$
\boxed{
fâ€™(c) = \frac{f(b)-f(a)}{b-a}
}
$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function and interval
f = lambda x: x**2
a, b = 1, 4

# Calculate average slope
avg_slope = (f(b) - f(a)) / (b - a)

# Find c (for this quadratic, derivative is 2x, so solve f'(c) = avg_slope)
c = avg_slope / 2

# Generate x values
x_vals = np.linspace(0, 5, 400)
y_vals = f(x_vals)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r'$f(x) = x^2$', color='blue')
plt.scatter([a, b], [f(a), f(b)], color='red', zorder=5)
plt.plot([a, b], [f(a), f(b)], '--', label='Secant line', color='purple')

# Tangent line at c
tangent_line = f(c) + (2*c)*(x_vals - c)
plt.plot(x_vals, tangent_line, '--', label=f'Tangent line at c={c:.2f}', color='green')
plt.scatter([c], [f(c)], color='green', s=80, zorder=5)

# Annotate
plt.title('Mean Value Theorem Illustration')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

## Extreme Value Theorem

If a function $f$ is continuous on a closed interval $[a, b]$ then $f$ attains both:
* an absolute maximum value, and
* an absolute minimum value

somewhere in the interval $[a, b]$.

Mathematically:

$
\exists\, c, d \in [a, b]\quad\text{s.t.}\quad
f(c) \leq f(x) \leq f(d)\quad \forall\, x \in [a, b]
$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define a continuous function on [a, b]
f = lambda x: np.sin(x) + 0.5*x
a, b = 0, 4

# Generate x values
x_vals = np.linspace(a, b, 400)
y_vals = f(x_vals)

# Find max and min values
max_index = np.argmax(y_vals)
min_index = np.argmin(y_vals)
x_max, y_max = x_vals[max_index], y_vals[max_index]
x_min, y_min = x_vals[min_index], y_vals[min_index]

# Plot the function and mark absolute extrema
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r'$f(x) = \sin(x) + 0.5x$', color='blue')
plt.scatter([x_max], [y_max], color='green', s=80, label='Absolute Maximum')
plt.scatter([x_min], [y_min], color='red', s=80, label='Absolute Minimum')

plt.title('Extreme Value Theorem Illustration')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

## Critical Numbers

A critical number of a function $f$ is a value $x = c$ in the domain of $f$ where:

1.	$fâ€™(c) = 0$  (derivative is zero â†’ horizontal tangent), or
2.	$fâ€™(c)$ does not exist, but $f(c)$ is defined.


## Concavity

Conditions for being concave downwards:
* slope is decreasing, $f'(x)$ is decreasing, i.e. the second derivative must be negative, $f''(x) < 0$.

Conditions for being concave upwards:
* slope is increasing, $f'(x)$ is increasing, i.e. the second derivative must be positive, $f''(x) > 0$.

### Inflection points:
* Inflection point is when the second derivative switches signs, i.e. this is the point when first derivative has a local minimum or maximum (switches the rate of growth).
* This is the point when the graph changes concavity.
* It can be the point when $g''(s) = 0$ and $g''(x)$ croses the $0$ line, or when $g''(x)$ is undefined.

```python
## Note on concavity and saddle points

import numpy as np
import matplotlib.pyplot as plt

# Define functions
x = np.linspace(-2, 2, 400)
f1 = x**3       # saddle with inflection
f2 = x**4       # flat minimum (no inflection)

# 2D saddle surface
X = np.linspace(-2, 2, 200)
Y = np.linspace(-2, 2, 200)
X, Y = np.meshgrid(X, Y)
Z = X**2 - Y**2  # hyperbolic paraboloid

# Plotting
plt.figure(figsize=(14, 4))

# 1D: x^3
plt.subplot(1, 3, 1)
plt.plot(x, f1, label=r'$f(x)=x^3$')
plt.title('Saddle point with inflection')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='red', linestyle='--', label='critical point')
plt.legend()
plt.grid(True)

# 1D: x^4
plt.subplot(1, 3, 2)
plt.plot(x, f2, label=r'$f(x)=x^4$')
plt.title('Flat minimum (no inflection)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='red', linestyle='--', label='critical point')
plt.legend()
plt.grid(True)

# 2D: Saddle
ax = plt.subplot(1, 3, 3, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm', alpha=0.9)
ax.set_title('2D Saddle Point')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.tight_layout()
plt.show()
```

Here are the visual examples:
1. Left: $f(x)=x^3$ â€“ a 1D saddle point with an inflection (concavity changes).
2. Middle: $f(x)=x^4$ â€“ a flat minimum, no concavity change, hence **not a saddle point**.
3. Right: A 2D saddle surface $z = x^2 - y^2$, showing how a saddle can exist without simple concavity in a single direction.


## Second Derivative Test of Concavity

* When $f'(c) = 0$, $f'$ exists in neigbourhood around $x = c$ while $f''(x)$ exists.
* Then, the second derivative determines,
1. $f''(c) < 0$ - function $f$ has a relative maximum value at $x = c$.
2. $f''(c) = 0$ - inconclusive.
3. $f''(c) > 0$ - function $f$ has a relative minimum value at $x = c$.

## Worked Example: optimization and sum of squares

What is the smallest possible sum of squares of two numbers, if their product is -16?

The relationships are,

$
S = x^2 + y^2
$

and,

$
x + y = -16
$

From this it follows that,

$
y = -\frac{16}{x}
$

Substituting this, we get updated sum of squares to,

$
S = x^2 + (-\frac{16}{x})^2 = x^2 + \frac{-16^2}{x^2}
$

The local extrema will need to satisfy $\frac{d}{dx} S = 0$ so that, the 

$
\frac{d}{dx} S = 2x + (-16)^2 \cdot (-2) \cdot x^{-3}
$

And, solving

$$
\begin{align}
2x + (-16)^2 \cdot (-2) \cdot x^{-3} &= 0 \\
2x &= - (-2) \cdot (-16)^2 \cdot x^{-3} \\
x & = (-16)^2 \cdot x^{-3} \\
(x^3) \cdot x &= (x^3) \cdot (-16)^2 \cdot x^{-3} \\
x^4 = (-16)^2 \\
x = \pm 4
\end{align}
$$

The local minimum will need to satisfy $\frac{d}{dx^2} S > 0$ at $x = \pm 4$. Therefore, evaluating the second derivative,

$
\frac{d}{dx^2} S = 2 + (-16)^2 \underbrace{\cdot (-2) \cdot(-3) \cdot x^{-4}}_{\text{always positive for any }x}
$

Since $\frac{d}{dx^2} S > 0$ then at any point the curve is bending downwards, so $x = \pm 4$ point to mimums.

Therefore, the solutions for $(x, y)$ are $\boxed{(4, -4)}$ or $\boxed{(-4, 4)}$ which produce the smallest possible sums of squares.
