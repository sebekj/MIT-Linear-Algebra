## Unit I: $\mathbf{Ax} = \mathbf{b}$ and the Four Subspaces

We write systems of linear equations in the matrix form $\mathbf{Ax} = \mathbf{b}$. We explore how the properties of $\mathbf{A}$ and $\mathbf{b}$ determine the solutions $\mathbf{x}$, if any exist. We pay particular attention to the solutions to $\mathbf{Ax} = 0$. And for a given matrix $\mathbf{A}$ we ask which $\mathbf{b}$ can be written in the form of $\mathbf{Ax}$.

### The Geometry of Linear Equations
(Source: https://ocw.mit.edu/courses/18-06sc-linear-algebra-fall-2011/pages/ax-b-and-the-four-subspaces/the-geometry-of-linear-equations/)

Three ways of looking at a problem of a set of linear equations (same number of equations and unknows)
- Row picture
- Column picture
- Matrix picture

*A worked example: Two equations and two unknowns.*

$$
\begin{align}
2x - y &= 0 \tag{1}\\
-x + 2y &= 3 \tag{2}
\end{align}
$$

This is the corresponding coeficient matrix of this system of linear equations:

$$
\begin{align}
\begin{bmatrix}
2 & -1 \\ -1 & 2
\end{bmatrix}
\begin{bmatrix}
x \\ y
\end{bmatrix} &=
\begin{bmatrix}
0 \\ 3
\end{bmatrix} \\
\\
\mathbf{Ax} &= \mathbf{b}
\end{align}
$$

Where $\mathbf{A}$ is the *coeficient matrix*, $\mathbf{x}$ is the vector of unknowns and $\mathbf{b}$ is a vector.

### The Row Picture
When I plot these linear equations, I obtain these two straight lines which intersect. The important question is whether a line goes through the point of origin.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for x-values
# Let's zoom in around 0 to 2, where the intersection (1,2) will be clearly visible
x = np.linspace(0, 2, 100)

# Express each line in slope-intercept form:
# 1) 2x - y = 0   => y = 2x
# 2) -x + 2y = 3  => 2y = x + 3 => y = x/2 + 3/2

y1 = 2*x
y2 = (x / 2) + (3 / 2)

# Plot the lines
plt.plot(x, y1, label="2x - y = 0  →  y = 2x")
plt.plot(x, y2, label="-x + 2y = 3 →  y = x/2 + 3/2")

# Intersection point (1, 2)
plt.scatter(1, 2, marker='o', zorder=5)  # 'zorder=5' to ensure it appears on top

# Label the axes and set a title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Lines and Their Intersection")

# Adjust axis limits to keep the intersection in clear view
plt.xlim(0, 2)
plt.ylim(0, 3)

plt.grid(True)
plt.legend()
plt.show()
```

### The Column Picture

In this picure, we are capturing the two equations at once.

$$
x
\begin{bmatrix}
2 \\ -1
\end{bmatrix} +y
\begin{bmatrix}
-1 \\ 2
\end{bmatrix} =
\begin{bmatrix}
0 \\ 3
\end{bmatrix}
$$

This equations asks us to combine vectors $\mathbf{x}$ and $\mathbf{y}$ in right amounts to get the $\mathbf{b}$. This is the *Linear Combination of the Columns* which is the most fundamental, in this course.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
x = np.array([2, -1])
y = np.array([-1, 2])
b = x + 2*y  # This should be [0, 3]

# Create a new figure
plt.figure()

# Draw coordinate axes
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# 1) Plot x (dotted arrow from origin)
plt.arrow(
    0, 0,          # start at origin
    x[0], x[1],    # end at x
    head_width=0.1, 
    length_includes_head=True, 
    linestyle=':', 
    color='blue',
    label='x = [2, -1]'
)

# We'll visualize 2y by adding y in two steps (tail-to-tip)

# 2) First y step: from x to x+y
start_point = x
mid_point = x + y
plt.arrow(
    start_point[0], start_point[1],
    y[0], y[1],
    head_width=0.1,
    length_includes_head=True,
    linestyle=':',
    color='red',
    label='y (step 1)'
)

# 3) Second y step: from (x+y) to x+2y
end_point = x + 2*y
plt.arrow(
    mid_point[0], mid_point[1],
    y[0], y[1],
    head_width=0.1,
    length_includes_head=True,
    linestyle=':',
    color='red',
    label='y (step 2)'
)

# 4) Plot b (solid arrow from origin)
plt.arrow(
    0, 0,
    b[0], b[1],
    head_width=0.1,
    length_includes_head=True,
    color='green',
    label='b = x + 2y = [0, 3]'
)

# Adjust axes
plt.xlim(-2.5, 3)
plt.ylim(-2, 3.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)

# Place legend in the bottom-left corner
plt.legend(loc='lower left')

plt.title("Visualizing b = x + 2y in Two Separate y Steps")
plt.show()
```

*Note:* All combinations of both vectors, multiplied $x$-times and $y$-times, would fill the whole plane.

Consider this $3 \times 3$ example, in three dimension $x, y, z$:

$$
\begin{align}
2x - y &= 0 \\
-x + 2y - z &= -1 \\
-3y + 4z &= 4
\end{align}
$$

The matrix picture:

$$
\mathbf{A} =
\begin{bmatrix}
2 & -1 & 0 \\ -1 & 2 & -1 \\ 0 & -3 & 4
\end{bmatrix}
\qquad
\mathbf{b} =
\begin{bmatrix}
0 \\ -1 \\ 4
\end{bmatrix}
$$

The row picture will show three planes which meet in one point. The row picture is getting hard to see.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Create a figure with 3D axes
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

# Define ranges for parameters
x_range = np.linspace(-3, 3, 50)
y_range = np.linspace(-3, 3, 50)
z_range = np.linspace(-3, 3, 50)

# --- Plane 1: 2x - y = 0  =>  y = 2x ---
# Parameterize by x and z (y is determined by x)
X1, Z1 = np.meshgrid(x_range, z_range)
Y1 = 2 * X1
surf1 = ax.plot_surface(X1, Y1, Z1, alpha=0.5, color='red', edgecolor='none')

# --- Plane 2: -x + 2y - z = -1  =>  z = -x + 2y + 1 ---
# Parameterize by x and y
X2, Y2 = np.meshgrid(x_range, y_range)
Z2 = -X2 + 2*Y2 + 1
surf2 = ax.plot_surface(X2, Y2, Z2, alpha=0.5, color='blue', edgecolor='none')

# --- Plane 3: -3y + 4z = 4  => y = 4(z - 1)/3
# Parameterize by y and z (y is determined by z)
X3, Z3 = np.meshgrid(x_range, z_range)
Y3 = (4 * Z3 - 4) / 3
surf3 = ax.plot_surface(X3, Y3, Z3, alpha=0.5, color='green', edgecolor='none')

# Label axes and set title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("Three Planes Defined by the Equations")

# Set equal aspect ratio (requires Matplotlib 3.3+)
ax.set_box_aspect([1, 1, 1])

# Adjust the view angle for a better perspective
ax.view_init(elev=20, azim=30)

# Create proxy artists for the legend since plot_surface doesn't support labels directly.
legend_elements = [
    Patch(facecolor='red', edgecolor='none', alpha=0.5, label=r'$2x-y=0$'),
    Patch(facecolor='blue', edgecolor='none', alpha=0.5, label=r'$-x+2y-z=-1$'),
    Patch(facecolor='green', edgecolor='none', alpha=0.5, label=r'$-3y+4z=4$')
]
ax.legend(handles=legend_elements, loc='lower left')

plt.show()
```

The column picture:

$$
x
\begin{bmatrix}
2 \\ -1 \\ 0
\end{bmatrix} + y
\begin{bmatrix}
-1 \\ 2 \\ -3
\end{bmatrix} + z
\begin{bmatrix}
0 \\ -1 \\ 4
\end{bmatrix} =
\begin{bmatrix}
0 \\ -1 \\ 4
\end{bmatrix}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # for older versions of matplotlib

# Define the three vectors (the "columns" in the linear combination)
v1 = np.array([2, -1, 0])
v2 = np.array([-1, 2, -3])
v3 = np.array([0, -1, 4])

# Right-hand side vector b
b = np.array([0, -1, 4])

# The given equation is: x*v1 + y*v2 + z*v3 = b
# In this case, the unique solution is: x = 0, y = 0, z = 1.
coeff_x = 0
coeff_y = 0
coeff_z = 1

# Compute the scaled vectors (their contributions)
v1_scaled = coeff_x * v1  # will be [0, 0, 0]
v2_scaled = coeff_y * v2  # will be [0, 0, 0]
v3_scaled = coeff_z * v3  # equals v3

# Set up a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

def plot_vector(ax, v, color, label):
    """Plot a 3D arrow from the origin and add a label at its tip."""
    ax.quiver(0, 0, 0, v[0], v[1], v[2],
              color=color, arrow_length_ratio=0.1)
    ax.text(v[0], v[1], v[2], label, color=color)

# Plot the three original (unscaled) vectors for reference
plot_vector(ax, v1, 'red', 'v1 = [2, -1, 0]')
plot_vector(ax, v2, 'blue', 'v2 = [-1, 2, -3]')
plot_vector(ax, v3, 'orange', 'v3 = [0, -1, 4]')

# Plot the right-hand side vector b
plot_vector(ax, b, 'green', 'b = [0, -1, 4]')

# Illustrate the tip-to-tail addition path for the scaled vectors:
# Start at the origin, add v1_scaled, then v2_scaled, then v3_scaled.
origin = np.array([0, 0, 0])
point_after_v1 = origin + v1_scaled   # still origin here
point_after_v2 = point_after_v1 + v2_scaled  # still origin
final_point = point_after_v2 + v3_scaled  # equals b

# Draw dotted lines to show the path
ax.plot([origin[0], point_after_v1[0]], [origin[1], point_after_v1[1]], [origin[2], point_after_v1[2]], 'k--')
ax.plot([point_after_v1[0], point_after_v2[0]], [point_after_v1[1], point_after_v2[1]], [point_after_v1[2], point_after_v2[2]], 'k--')
ax.plot([point_after_v2[0], final_point[0]], [point_after_v2[1], final_point[1]], [point_after_v2[2], final_point[2]], 'k--')

# Set axis limits based on all vectors for a balanced view
all_points = np.array([v1, v2, v3, b])
max_range = np.max(np.abs(all_points)) + 1
ax.set_xlim([-max_range, max_range])
ax.set_ylim([-max_range, max_range])
ax.set_zlim([-max_range, max_range])

# Label axes and set an equal aspect ratio
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_box_aspect([1, 1, 1])
ax.view_init(elev=20, azim=30)
ax.set_title("Linear Combination: x*v1 + y*v2 + z*v3 = b")

plt.show()
```

This equations asks for the right combination which is $x = 0$, $y = 0$ and $z = 1$. The big picture is that in the row picture I have three planes, in the column pictures, I combine the columns to receive the right hand column.

Can I solve $\mathbf{Ax} = \mathbf{b}$ for every right handside  $\mathbf{b}$? Is there a solution? If there is then elimination will give the solution. Do the linear combinations of the columns fill three dimensional space (i.e. every $\mathbf{b}$). The answer for this particular matrix is yes. This is a non-singular, invertible matrix. There may be other matrices when the answer is no.

How could it go wrong? When the three columns lie in the same plane, then their combinations lie in the same plane while $\mathbf{b}$ is in a different plane. E.g. when vector $\mathbf{x}$ is just a sum of vector $\mathbf{y}$ and $\mathbf{z}$. This would solve it for some right hand $\mathbf{b}$'s. The matrix would be *not invertible*.

Imagine if have nine dimension, vectors with nine components. If I choose the columns so that they are not independent, then I will not be able to combine them to achieve all right-handside $\mathbf{b}$'s.

How do you multiply a matrix with a vector?

Using the column view:

$$
\begin{aligned}
\mathbf{Ax} &= \mathbf{b} \\
\\
\begin{bmatrix}
2 & 5 \\
1 & 3
\end{bmatrix}
\begin{bmatrix}
1 \\
2
\end{bmatrix}
&=
1 \cdot
\begin{bmatrix}
2 \\
1
\end{bmatrix}
+
2 \cdot
\begin{bmatrix}
5 \\
3
\end{bmatrix}
=
\begin{bmatrix}
12 \\
7
\end{bmatrix}
\end{aligned}
$$

Here, this method shows that $\mathbf{Ax}$ is a combination of columns of $\mathbf{A}$.

You may also calculate the product $\mathbf{Ax}$ by taking the *dot product* of each row of $\mathbf{A}$ with the vector $\mathbf{x}$:

$$
\begin{align}
\begin{bmatrix}
2 & 5 \\
1 & 3
\end{bmatrix}
\begin{bmatrix}
1 \\ 2
\end{bmatrix} =
\begin{bmatrix}
2 \cdot 1 + 5 \cdot 2 \\
1 \cdot 1 + 3 \cdot 2
\end{bmatrix} =
\begin{bmatrix}
12 \\ 7
\end{bmatrix}
\end{align}
$$


In a *singular matrix*, the matrix of coefficients cannot solve all possible vectors $\mathbf{b}$ because the column vectors are *lineary dependent*. All linear combinations of such dependent vectors will lie on a point or line (in two dimensions) or on a point, line or plane (in three dimensions). The combinations do not fill the whole space.

### Reading

Reading *Introduction to Linear Algebra* by Gilbert Strang. Sections 1.1., 1.2, 2.1.

- **The column space of a matrix $\mathbf{A}$** with two columns is a plane in three-dimensional space. For any $\mathbf{A}$, the column space of $\mathbf{A}$ contains all combinations of the columns.
- From left to right, each column in a matrix can be **independent** of the previous columns or it can be a **combination** of those columns.
- **The linear combinations** of vectors $\mathbf{v} \begin{bmatrix} v_1 \\ v_2 \end{bmatrix}$ and $\mathbf{w} \begin{bmatrix} w_1 \\ w_2 \end{bmatrix}$ fill the $xy$ plane.
- **The length of vector** of vector $\mathbf{w}$ is $\left\| \mathbf{w} \right\| = \sqrt{w^2_1 + w^2_2}$.
- Algebraically, **The dot product** of vectors $\mathbf{v}$ and $\mathbf{w}$, each with two components, is $\mathbf{v} \cdot \mathbf{w} = v_1w_1 +v_2w_2$.
- Linear combinations of vectors $\mathbf{v}$ and $\mathbf{w}$ are built from **multiplication** and **addition**.
- A $2 \times 2$ matrix $\mathbf{A} = \begin{bmatrix} \mathbf{v} \quad \mathbf{w} \end{bmatrix}$ is **invertible** when the vectors $\mathbf{v}$ and $\mathbf{w}$ are **linearly independent**, i.e. their linear combinations fill the $xy$ plane.
- When multiplying any vector with an **identity matrix** that vector remains unchanged; it is analog to multiplying by $1$ in $\mathbb{R}$ (real numbers) or $\mathbb{C}$ (complex numbers) fields.
- This is the generalisation for the $n \times n$ **identity matrix**:

$$
\mathbf{I}_n =
\begin{bmatrix}
1 & 0 & \cdots & 0 \\ 0 & 1 & \cdots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \cdots & 1
\end{bmatrix}
$$

- When two vectors are considered **independent**, their arrows point in different directions.
- For two vectors, you can check whether one is a scalar multiple of the other if all corresponding pairs have the same ratio (non zero):

$$
\mathbf{u} =
\begin{bmatrix}
u_1 \\ u_2 \\ \vdots \\ u_n
\end{bmatrix}
\quad \text{and} \quad
\mathbf{v} =
\begin{bmatrix}
v_1 \\ v_2 \\ \vdots \\ v_n
\end{bmatrix}
$$

That is:

$$
\frac{v_1}{u_1} = \frac{v_2}{u_2} = \cdots = \frac{v_n}{u_n}
$$

Now, no denominator is $0$ or both denominator and numerator are $0$ at the same time as scalar multiple $m$ of $0$ is always $0$. The idea is that exists a scalar $m$ such that $v_i = mu_i$ for all $i$.

**Geometrically**, the dot product measures how much one vector extends in the direction of another. It is a measure of similarity of direction. It tells how much one vector 'projects' onto another. It is defined by:

$$
\mathbf{u} \cdot \mathbf{v} = \lVert \mathbf{u} \rVert \lVert \mathbf{v} \rVert \cos \theta
$$

Intuitively:
- **Alignment**: If the vectors point in a similar direction, $\theta$ is small and $\cos \theta$ is close to $1$, so the dot product is large and positive.
- **Orthogonality**: if the vectors are perpendicular, $\theta = 90^\circ$ and $\cos 90^\circ = 0$, so the dot product is $0$. This tells that the vectors have no component in the direction of each other.
- **Opposite Directions**: If the vectors point in the opposite directions, $\theta$ is close to $180^\circ$ and $\cos\theta$ is negative, so the dot product is negative.

**Algebraically**, the dot product $\mathbf{u} \cdot \mathbf{v}$ is defined:

$$
\mathbf{u} \cdot \mathbf{v} = u_1v_1 + u_2v_2 + \cdots + u_nv_n
$$

Intuitions:
- **Component-wise Contribution**: When you calculate the dot product algebraically, each term $u_iv_i$ represents how much the corresponding components of $\mathbf{v}$ contribute in the direction of $\mathbf{u}$. If the two vectors point in the similar directions, their corresponding components have the same sign, and the sum is large. If they are perpendicular, the contributions cancel out ina way that yields zero.
- **Extracting the Projection Length**: By dividing the dot product $\mathbf{u} \cdot \mathbf{v}$ by $\lVert \mathbf{u} \rVert$, you effectivelly 'cancel out' the effect of $\mathbf{u}$'s magnitude, leaving only the length of $\mathbf{v}$ that is aligned with $\mathbf{u}$. This is the projection of length.

**Unit Vector**:
- A unit vector $\mathbf{u}$ has length $\lVert \mathbf{u} \rVert = 1$.
- If $\mathbf{v} \neq 0$ then this is the unit vector:

$$
\mathbf{u} = \frac{\mathbf{v}}{\lVert \mathbf{v} \rVert}
$$

*Note: the unit vector $\mathbf{u}$ is the vector of length one along the original vector placed at the origin of vector $\mathbf{v}$. The unit vector is* dependent *on the original vector and the scalar is the* inverse of length *of the original vector, that is $\frac{1}{\lVert \mathbf{v} \rVert}$.*

- The standard unit vector along the $x$ axis is written $\mathbf{i}$.
- In the $xy$ plane, the unit vector that makes an angle $\theta$ with the $x$ axis is $\mathbf{u} = (\cos\theta, \sin\theta)$:

$$
\text{Unit vectors} \quad \mathbf{i} = \begin{bmatrix} 1 \\ 0 \end{bmatrix} \quad \text{and} \quad
\mathbf{u} = \begin{bmatrix} \cos\theta \\ \sin\theta \end{bmatrix}. \quad \text{Notice} \quad
\mathbf{i} \cdot \mathbf{u} = \cos\theta
$$

- $\mathbf{u} = (\cos\theta, \sin\theta)$ is a unit vector because $\mathbf{u} \cdot \mathbf{u} = \cos^2 \theta + \sin^2 \theta = 1$
- Perpendicular vectors have $\mathbf{v} \cdot \mathbf{w} = 0$. Then $\lVert \mathbf{v} + \mathbf{w} \rVert^2 = \lVert \mathbf{v} \rVert^2 + \lVert \mathbf{w} \rVert^2$.
- Algebraically, this special case when $\mathbf{v} \cdot \mathbf{w} = 0$:

$$
\lVert \mathbf{v} + \mathbf{w} \rVert^2 = (\mathbf{v} + \mathbf{w}) \cdot (\mathbf{v} + \mathbf{w}) = \mathbf{v} \mathbf{v} + \mathbf{v} \mathbf{w} + \mathbf{w} \mathbf{v} + \mathbf{w} \mathbf{w} = \lVert \mathbf{v} \rVert^2 + \lVert \mathbf{w} \rVert^2
$$

- Note that two of the terms equal zero ($\mathbf{v} \cdot \mathbf{w} = 0$). But also, the same is true for:

$$
\lVert \mathbf{v} - \mathbf{w} \rVert^2 = \lVert \mathbf{v} \rVert^2 + \lVert \mathbf{w} \rVert^2
$$

### Unit Circle

```python
import matplotlib.pyplot as plt
import numpy as np

# Set up figure
fig, ax = plt.subplots(figsize=(6,6))

# Unit circle
theta = np.linspace(0, 2*np.pi, 300)
x = np.cos(theta)
y = np.sin(theta)
ax.plot(x, y, 'lightgray', label='Unit Circle')

# Vectors
i = np.array([1, 0])  # formerly vector v
theta_angle = np.pi / 4  # 45 degrees
u = np.array([np.cos(theta_angle), np.sin(theta_angle)])

# Dot product to get projection length
proj_length = np.dot(u, i) / np.linalg.norm(i)**2  # = cos(theta) since i is unit
proj_vector = proj_length * i  # projection of u onto i

# Plot vectors
ax.quiver(0, 0, i[0], i[1], angles='xy', scale_units='xy', scale=1,
          color='blue', label=r'$\mathbf{i} = (1, 0)$')
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1,
          color='orange', label=r'$\mathbf{u} = (\cos\theta, \sin\theta)$')

# Plot projection of u onto i
ax.quiver(0, 0, proj_vector[0], proj_vector[1], angles='xy', scale_units='xy', scale=1,
          color='green', linestyle='-', linewidth=2, label=r'$\mathrm{proj}_{\mathbf{i}} \mathbf{u}$')

# Dashed line from tip of projection to tip of u
ax.plot([proj_vector[0], u[0]], [proj_vector[1], u[1]], 'k--', linewidth=1)

# Add labels for cos(theta) and sin(theta)
ax.text(proj_vector[0]/2 - 0.1, proj_vector[1] - 0.08, r'$\cos\theta$', color='green', fontsize=12)
ax.text(u[0] + 0.05, u[1]/2, r'$\sin\theta$', color='black', fontsize=12)

# Angle arc
arc_theta = np.linspace(0, theta_angle, 100)
arc_radius = 0.3
arc_x = arc_radius * np.cos(arc_theta)
arc_y = arc_radius * np.sin(arc_theta)
ax.plot(arc_x, arc_y, 'gray')
ax.text(0.35, 0.1, r'$\theta$', fontsize=14)

# Axis and format
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect('equal')
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.set_xticks([])
ax.set_yticks([])
ax.legend(loc='upper left')
ax.set_title('Projection of $\mathbf{u}$ onto $\mathbf{i}$')

plt.show()
```

### A Sidestep: a take on $\sin$ and $\cos$, an intuition

*Q: How do the coordinates of vector $\mathbf{u}$ turn out to be ($\cos\theta$,$\sin\theta$)?*

It follows from the definition of the 

*Note on the Unit Circle:*
- The **Unit Circle** has a radius equal to $1$.
- Here, $\text{proj}_{\mathbf{i}}\mathbf{u}$, the projection of vector $\mathbf{u}$ onto $\mathbf{i}$, is $\cos \theta$, i.e. it is the *cosine* of the angle $\theta$ between the two vectors.
- The distance betwen point $(\sin\theta, \cos\theta$), which is on the Unit Circle, and vector $\mathbf{i}$ is $\sin\theta$.

### Angle Between Two Vectors
The **dot product** $\mathbf{v} \cdot \mathbf{w}$ finds the angle between any two nonzero vectors $\mathbf{v}$ and $\mathbf{w}$.

Consider unit vectors $\mathbf{v} = (\cos\alpha, \sin\alpha)$ and $\mathbf{w} = (\cos\beta, \sin\beta)$. They have a dot product $\mathbf{v} \cdot \mathbf{w} = \cos\alpha\cos\beta + \sin\alpha\sin\beta$. In trigonometry, this is the formula for $\cos(\beta - \alpha)$. The angle between the unit vectors $\mathbf{v}$ and $\mathbf{w}$ is $\beta - \alpha$. Now, $\theta = \beta - \alpha$.

*The Insight:*
- The angle between two unit vectors on the unit circle is just the difference of their angles.
- The dot product of those vectors equals the cosine of that angle difference.
- The angles $\alpha$ and $\beta$ are measured from the positive $x$-axis.

**The dot product reveals the exact angle $\theta$**. For unit vectors $\mathbf{u}$ and $\mathbf{U}$ the dot product $\mathbf{u} \cdot \mathbf{U}$ is the cosine of $\theta$. This remains true in $n$ dimensions.

- The unit vectors $\mathbf{v} = (1, 0)$ and $\mathbf{u} = (\cos\theta, \sin\theta)$ have $\mathbf{v} \cdot \mathbf{u} = \cos \theta$. Connecting the dot product to the angle between vectors: If $\lVert \mathbf{v} \rVert = 1$ and $\lVert \mathbf{u} \rVert = 1$, the angle $\theta$ beetween $\mathbf{v}$ and $\mathbf{u}$ has $\cos\theta = \mathbf{v} \cdot \mathbf{u}$.
- The dot product reveals the exact angle $\theta$ which is valid in $n$ dimensions. The $\cos \theta$ for unit vectors $\mathbf{u}$ and $\mathbf{U}$ is revealed by the dot product $\mathbf{u} \cdot \mathbf{U}$.
- If vectors $\mathbf{v}$ and $\mathbf{w}$ are not unit vectors, it follows from $\mathbf{u} = \frac{\mathbf{v}}{\lVert \mathbf{v} \rVert}$ and $\mathbf{U} = \frac{\mathbf{w}}{\lVert \mathbf{w} \rVert}$ this **cosine formula** for angle $\theta$:

$$
\frac{\mathbf{v} \cdot \mathbf{w}}{\lVert \mathbf{v} \rVert \lVert \mathbf{w} \rVert} = \cos \theta
$$

### Schwarz Inequality

Whatever the angle, this dot product of $\mathbf{v}/\lVert \mathbf{v} \rVert$ with $\mathbf{w}/\lVert \mathbf{w} \rVert$ never exeeds one, it ranges from $-1$ to $1$ (the absolute value ranges from $0$ to $1$).

Starting from the *cosine formula* we take the absolute value of both sides:

$$
\left| \frac{\mathbf{v} \cdot \mathbf{w}}{\lVert \mathbf{v} \rVert \lVert \mathbf{w} \rVert} \right| = |\cos\theta| \le 1
$$

Now, we multiply both sides with $\lVert \mathbf{v} \rVert \lVert \mathbf{w} \rVert$ (which are non-negative):

$$
|\mathbf{v} \cdot \mathbf{w}| \le \lVert \mathbf{v} \rVert \lVert \mathbf{w} \rVert
$$

*Intuition:*

- The dot product measures how aligned two vectors are.
- the product $\lVert \mathbf{v} \rVert \lVert \mathbf{w} \rVert$ is the **maximum possible value** of the dot product--it occurs when the vectors point in exactly the same or opposite directions, i.e. when $|\cos\theta| = 1.
- For any other angle, the dot product is *less than* that maximum.

This is the **Schwarz inequality** $|\mathbf{v} \cdot \mathbf{w}| \le \lVert \mathbf{v} \rVert \lVert \mathbf{w} \rVert$ for dot products. From this the **Triangle Inequality** can be derived:

1. Start with the definition of the squared norm (length):

$$
\lVert \mathbf{v} + \mathbf{w}\rVert^2 = (\mathbf{v} + \mathbf{w}) \cdot (\mathbf{v} + \mathbf{w})
$$

2. Now, expand the dot product

$$
= \mathbf{v} \cdot \mathbf{v} + 2 \mathbf{v} \cdot \mathbf{w} + \mathbf{w} \cdot \mathbf{w}
$$

3. This simplifies to:

$$
= \lVert \mathbf{v} \rVert^2 + 2\mathbf{v}\cdot\mathbf{w} + \lVert \mathbf{w} \rVert^2
$$

4. Now apply Schwarz inequality:

$$
\mathbf{v} \cdot \mathbf{w} \le \lVert \mathbf{v} \rVert \lVert \mathbf{w} \rVert
$$

5. So the upper bound:

$$
\lVert \mathbf{v} + \mathbf{w} \rVert^2 \le \lVert \mathbf{v} \rVert^2 + 2\lVert \mathbf{v} \rVert \lVert \mathbf{w} \rVert^2 + \lVert \mathbf{w} \rVert^2
$$

6. Which is:

$$
= (\lVert \mathbf{v} \rVert + \lVert \mathbf{w} \rVert)^2
$$

7. Now, take the wquare roots of both sides (both are non-negative):

$$
\lVert \mathbf{v} + \mathbf{w} \rVert \le \lVert \mathbf{v} \rVert + \lVert \mathbf{w} \rVert
$$

This is the **Triangle Inequality**.

### Triangle Inequality
The **Triangle Inequality** $\lVert \mathbf{v} + \mathbf{w} \rVert \le \lVert \mathbf{v} \rVert + \lVert \mathbf{w} \rVert$ means that the length of the sum of two vectors is less than or equal to the sum of their lengths. The triangle inequality comes directly from Schwarz Inequality.
